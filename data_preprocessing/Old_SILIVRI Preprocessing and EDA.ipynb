```python
def create_dirpath_list(root_dir):
    dirpath_list=[]
    for dirpath, dirnames, filenames in os.walk(root_dir):
        dirpath_list.append(dirpath)
    dirpath_list=[Path(p) for p in dirpath_list]
    return dirpath_list


dirpath_list=create_dirpath_list(prepared_wfs_daily_path)
```

```python
def create_savepath_list(dirpath_list,old_base,new_base): 
    savepath_list = [new_base / path.relative_to(old_base) for path in dirpath_list]
    for i in savepath_list:
        i.parent.mkdir(parents=True, exist_ok=True) 
    return savepath_list
```

```python
def fast_walk(path):
    with os.scandir(path) as it:
        for entry in it:
            if entry.is_dir(follow_symlinks=False):
                yield from fast_walk(entry.path)
            elif entry.is_file(follow_symlinks=False):
                yield entry.name
```
